```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1]) #LARGE FIGURE
axes2 = fig.add_axes([0.2,0.2,0.5,0.5]) # SMALL FIGURE

axes1.plot(a,b)

axes1.set_xlabel("X Lable")
axes1.set_ylabel("Y Lable")
axes1.set_title("Big Figure")

axes2.plot(a,b)
axes2.set_title("Small Figure")
plt.show()
```

```python
# Creates blank canvas
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1]) # Large figure
axes2 = fig.add_axes([0.2,0.5,0.25,0.25]) # Small figure

axes1.plot(a,b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In')

plt.show()
```

```python
# Creates blank canvas
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1]) # Large figure
axes2 = fig.add_axes([0.2,0.5,0.25,0.25]) # Small figure
axes3 = fig.add_axes([1, 1, 0.25, 0.25]) # Starts at top right corner!

axes1.plot(a,b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In')

axes3.plot(a,b)

plt.show()
```

```python
fig = plt.figure(figsize= (12,8),dpi = 200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
plt.show()
```
